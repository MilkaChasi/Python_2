### Manipulación del data frame

Se realiza una manipulación y transformación de los datos para lograr crear una medición de tendencia central y disperción de la felicidad a nivel mundial para el primer y ultimo año. Por lo cual se aplica la opción **groupby()**; permite agrupar los datos de un data frame por los valores de una columna. En este caso, agrupar los datos por los valores de la columna *pais*. Para calcular el promedio se aplica **mean()** y para desviación standar se aplica **std()**, utilizando los valores de la columna *escalera_vida* para el calculo.

```python
pais_prom = datosfelicidad.groupby('anio')['escalera_vida'].mean()
pais_prom
```

```python
pais_std = datosfelicidad.groupby('anio')['escalera_vida'].std()
pais_std
```

Como se observa en el paso anterior se devuelve una serie de pandas con los valores promedio de la columna *escalera_vida* para cada país. Verificar con la función type.

```python
type(pais_prom)
```

```python
type(pais_std)
```

Se convierte estas series en un data frame usando el método **to_frame()**. Este método recibe como argumento el nombre de la columna que se quiere asignar al data frame.

```python
pais_prom = pais_prom.to_frame()
pais_prom
```

```python
pais_std = pais_std.to_frame()
pais_std
```

Se observa los nombres de las columnas están en dos niveles diferentes. El primer nivel es el nombre de la columna que se definio más arriba, mientras que el segundo nivel es el índice de las filas. Verificar que esto es cierto usando el método **index**.

```python
pais_prom.index
```

```python
pais_std.index
```

Para que pais aparezca como una columna en vez de un índice, se utiliza el método **reset_index()**. Este método devuelve un data frame con el índice de las filas como una columna.

```python
pais_prom.reset_index() 
```

```python
pais_std.reset_index() 
```

Para evitar tener que asignar el resultado a una variable, se utiliza el parámetro **inplace=True** para que el cambio se haga directamente en el data frame original.

```python
pais_prom.reset_index(inplace = True)
```

```python
#Asignación de nombre a la  nueva columna
pais_prom = pais_prom.rename(columns={'escalera_vida':'promedio'})
```

```python
pais_prom
```

```python
pais_std.reset_index(inplace = True)
pais_std
```

Para eliminar una columna se utiliza el metodo **drop()**. Su eliminación se debe a que se requerirá solo de la columna desviación estandar en los próximos pasos. 

```python
pais_std = pais_std.drop(['anio'], axis=1)
```

```python
#Asignación de nombre a la nueva columna
pais_std = pais_std.rename(columns={'escalera_vida':'desviacion_standr'})
```

```python
pais_std
```

Ahora se utiliza el método **concat()** para unir los dos data frames por columnas. Recordar especificar los data frames que se necesita unir como una lista.

```python
medidafelicidad = pd.concat([pais_prom, pais_std], axis = 1)
```

```python
medidafelicidad
```

```python
años_medidafelicidad = medidafelicidad[(medidafelicidad['anio'] == 2005) | (medidafelicidad['anio'] == 2018)]
años_medidafelicidad
```
